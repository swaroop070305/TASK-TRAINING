**Conditional statements**

if Statement:

```python
# if statement

age = 20

if age >= 18:
    print("You are eligible to vote.")
```

if else statement:

```python
# if else statement

num = 7

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

elif Statement:


```python
# elif statement

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

Nested if Statement:

```python
# nested if statement

age = 20
has_license = True

if age >= 18:
    if has_license:
        print("You can drive.")
    else:
        print("Get a driving license first.")
else:
    print("You are not eligible to drive.")
```

 **Looping Statements**




for Loop:

```python
# for loop

for i in range(1, 6):
    print(i)
```

while Loop:

```python
# while loop

count = 1

while count <= 5:
    print(count)
    count += 1
```
