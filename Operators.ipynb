Arithmetic Operators

```python
# Arithmetic Operators

a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
```

Assignment Operators

```python
# Assignment Operators

a = 5
print(a)

a += 3
print(a)

a *= 2
print(a)
```

Logical Operators

```python
# Logical Operators

a = True
b = False

print(a and b)
print(a or b)
print(not a)
```

Relational Operators

```python
# Relational Operators

a = 15
b = 10

print("a =", a)
print("b =", b)

print("a == b :", a == b)
print("a != b :", a != b)
print("a > b  :", a > b)
print("a < b  :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
```
