# Functions in Python



## Simple Function

```python
# Function without parameters

def greet():
    print("Hello, Welcome!")

greet()
```

## Function with Parameters

```python
# Function with parameters

def add(a, b):
    print("Sum =", a + b)

add(10, 20)
```

## Function with Return Value

```python
# Function with return value

def square(n):
    return n * n

result = square(5)
print(result)
```

## Default Parameter

```python
# Default parameter

def student(name="Rahul"):
    print("Name:", name)

student()
student("Ravi")
```

## Keyword Arguments

```python
# Keyword arguments

def details(name, age):
    print(name, age)

details(age=20, name="Anu")
```

## Arbitrary Arguments (*args)

```python
# *args example

def total(*numbers):
    print(sum(numbers))

total(10, 20, 30, 40)
```

## Lambda Function

```python
# Lambda function

square = lambda x: x * x

print(square(6))
```

## Recursive Function

```python
# Recursive function

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```
