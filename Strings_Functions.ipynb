# Python Strings



## Create a String

```python
text = "Hello Python"
print(text)
print(type(text))
```

## Length

```python
text = "Hello Python"
print(len(text))
```

## Upper

```python
text = "Hello Python"
print(text.upper())
```

## Lower

```python
text = "Hello Python"
print(text.lower())
```

## Title

```python
text = "hello python"
print(text.title())
```

## Capitalize

```python
text = "hello python"
print(text.capitalize())
```

## Replace

```python
text = "Hello Python"
print(text.replace("Python","World"))
```

## Find

```python
text = "Hello Python"
print(text.find("Python"))
```

## Count

```python
text = "banana"
print(text.count("a"))
```

## Startswith

```python
text = "Hello Python"
print(text.startswith("Hello"))
```

## Endswith

```python
text = "Hello Python"
print(text.endswith("Python"))
```

## Strip

```python
text = "  Python  "
print(text.strip())
```

## Split

```python
text = "Python is easy"
print(text.split())
```

## Join

```python
words = ["I","love","Python"]
print(" ".join(words))
```

## Indexing

```python
text = "Python"
print(text[0])
print(text[-1])
```

## Slicing

```python
text = "Python"
print(text[1:4])
```

## Isalpha

```python
text = "Python"
print(text.isalpha())
```

## Isdigit

```python
text = "12345"
print(text.isdigit())
```

## Isalnum

```python
text = "Python123"
print(text.isalnum())
```

## Swapcase

```python
text = "PyThOn"
print(text.swapcase())
```
