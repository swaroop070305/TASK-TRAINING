# Python Collections



## List

```python
# List

fruits = ["Apple", "Banana", "Mango"]

print(fruits)
print(type(fruits))
print(fruits[0])
```

## Tuple

```python
# Tuple

numbers = (10, 20, 30)

print(numbers)
print(type(numbers))
print(numbers[1])
```

## Set

```python
# Set

colors = {"Red", "Blue", "Green"}

print(colors)
print(type(colors))
```

## Dictionary

```python
# Dictionary

student = {
    "Name": "Rahul",
    "Age": 20,
    "Branch": "ECE"
}

print(student)
print(type(student))
print(student["Name"])
```
