# Data Structures in Python



## Stack (LIFO)

```python
# Stack using list

stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print("Stack:", stack)

print("Popped:", stack.pop())
print("After Pop:", stack)
```

## Queue (FIFO)

```python
# Queue using list

queue = []

queue.append(10)
queue.append(20)
queue.append(30)

print("Queue:", queue)

print("Removed:", queue.pop(0))
print("After Remove:", queue)
```

## Linked List

```python
# Simple Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

first = Node(10)
second = Node(20)
third = Node(30)

first.next = second
second.next = third

temp = first

while temp:
    print(temp.data)
    temp = temp.next
```

## Tree

```python
# Simple Binary Tree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = Node(10)
root.left = Node(20)
root.right = Node(30)

print("Root:", root.data)
print("Left Child:", root.left.data)
print("Right Child:", root.right.data)
```
