# Classes in Python



## 1. Simple Class

```python
# Simple Class

class Student:
    name = "Rahul"

s = Student()

print(s.name)
```

## 2. Class with Constructor

```python
# Constructor

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s = Student("Rahul", 20)

print(s.name)
print(s.age)
```

## 3. Class Method

```python
# Method in a class

class Student:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)

s = Student("Anu")
s.display()
```

## 4. Another Example

```python
# Calculator class

class Calculator:
    def add(self, a, b):
        return a + b

c = Calculator()

print(c.add(10, 20))
```
