```python
import numpy as np
import pandas as pd
s=pd.Series((1,2,6,8,9,4,7))
print(s)
print(s.describe())
print(s.mean())
print(s.std())
print(s.min())
print(s.max())
print(s.map(lambda x:x*2))
print(s.apply(lambda x:x**4))
print(s.sort_values())
print(s.sum())
print(s.cumsum())
print(s.aggregate(['sum','mean','std']))
```

```python
import numpy as np
import pandas as pd
s=pd.Series((1,2,6,np.nan,9,4,7))
print(s)
print(s.iloc[-2])
print(s.loc[2])
print(s[s>5])
```
