# Plotly



## Import Plotly

```python
import plotly.express as px

print("Plotly imported successfully")
```

## Create Sample Data

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Anu", "Ravi", "Priya"],
    "Marks": [85, 90, 78, 92],
    "Age": [20, 21, 20, 22]
}

df = pd.DataFrame(data)
print(df)
```

## Bar Chart

```python
fig = px.bar(df, x="Name", y="Marks", title="Student Marks")
fig.show()
```

## Line Chart

```python
fig = px.line(df, x="Name", y="Marks", title="Marks Line Chart")
fig.show()
```

## Scatter Plot

```python
fig = px.scatter(df, x="Age", y="Marks", color="Name",
                 title="Age vs Marks")
fig.show()
```

## Pie Chart

```python
fig = px.pie(df, names="Name", values="Marks",
              title="Marks Distribution")
fig.show()
```

## Histogram

```python
fig = px.histogram(df, x="Marks",
                    title="Marks Histogram")
fig.show()
```
