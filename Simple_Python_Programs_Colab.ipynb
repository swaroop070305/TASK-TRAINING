# Simple Python Programs



## Prime Number

```python
num = int(input("Enter a number: "))

count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

## Factorial

```python
num = int(input("Enter a number: "))

fact = 1
for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```

## Reverse of String

```python
text = input("Enter a string: ")

print("Reverse =", text[::-1])
```

## Reverse of Number

```python
num = int(input("Enter a number: "))

rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reverse =", rev)
```

## Palindrome

```python
text = input("Enter a word: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

## Armstrong Number

```python
num = int(input("Enter a number: "))

temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** 3
    temp = temp // 10

if total == num:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

## Fibonacci Series

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

## Maximum Number

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print(a, "is greater")
else:
    print(b, "is greater")
```

## Swap Without Third Variable

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a = a + b
b = a - b
a = a - b

print("a =", a)
print("b =", b)
```
