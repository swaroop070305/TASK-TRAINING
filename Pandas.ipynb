## Create a DataFrame

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Anu", "Ravi"],
    "Age": [20, 21, 22],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)

print(df)
```

## Display Rows and Columns

```python
print(df.head())
print()
print(df.tail())
```

## Display Column

```python
print(df["Name"])
```

## Add a New Column

```python
df["Grade"] = ["A", "A+", "A"]

print(df)
```

## Basic Information

```python
print(df.info())
```

## Statistics

```python
print(df.describe())
```

## Filter Data

```python
print(df[df["Marks"] > 85])
```

## Sort Values

```python
print(df.sort_values("Marks"))
```
